```python
import pandas as pd
import numpy as np
import csv

df = pd.read_csv('/content/merged_dataset.csv')
df.head(10)
```

```python
nan_value = float("NaN")
df.replace("", nan_value, inplace=True)
df.dropna(subset = ["wind_speed"], inplace=True)
```

```python
df.head(10)
```

```python
df.to_csv("cleaned_data.csv", encoding='utf-8')
```

```python
# Cleaning the zero values
coordinates = []
for index, row in df.iterrows():
  if (row['humidity'] ==0):
    coordinates.append([row['Latittude'],row['Longitude'],index])

print(coordinates)
```

```python
print(len(coordinates))
```

```python
pressure_df = pd.read_csv('/content/pressure_humidity_tem.csv')
pressure_df.head(10)
```

```python
pressure_array =[0]*116
humidity_array = [0]*116;
temparature =[0]*116;

i=0
for coord in coordinates:
  print(coord)
  wind_lat = coord[0]
  wind_long = coord[1]
  for index1, pressure_row in pressure_df.iterrows():
    pre_lat = pressure_row['latitude']
    pre_long = pressure_row['longitude']

    if ((wind_lat-10)<pre_lat<(wind_lat+10)):
      print("Found lattitude values")
      print("lat :",pre_lat,"Index :",index1)
      
      #Check longitude
      if ((wind_long-25)<pre_long<(wind_long+25)):
        print("Found longitude values")
        pressure_array[i] = pressure_row['surface_pressure']
        humidity_array[i] = pressure_row['specific_humidity']
        temparature[i] = pressure_row['temperature']
        break
        
    else:
      print("In else")
  i+=1

print(pressure_array)
```

```python
copy_df = df.copy()
# Edit 
i =0
for coord in coordinates:
  ind = coord[2]
  copy_df.at[ind,'pressure'] = pressure_array[i]
  copy_df.at[ind,'humidity'] = humidity_array[i]
  copy_df.at[ind,'temperature'] = temparature[i]

  i+=1
```
