```python
import pandas as pd
import numpy as np
import csv


df=pd.read_csv('/content/trimmed-August2021_csv.csv',error_bad_lines = False)
df.head(10)

```

```python
# Getting a smaller dataset since it is exhaustive
print(" **** INFO ******",df.info())
```

```python
# Selecting 5000 only
smaller_df = df[0:6001]
smaller_df.head(10)
```

```python

wind_lat = 14.777475	
wind_long = 291.68790

pressure_df = pd.read_csv('/content/pressure_humidity_tem.csv')
pressure_df.head(10)
```

```python
pressure_array =[0]*16384
humidity_array =[0]*163684
temparature = [0]*163684

# Iterate over rows and columns
for index, row in smaller_df.iterrows():
  wind_lat = row['Latittude']
  wind_long = row['Longitude']
  for index1, pressure_row in pressure_df.iterrows():
    pre_lat = pressure_row['latitude']
    pre_long = pressure_row['longitude']

    if ((wind_lat-2)<pre_lat<(wind_lat+2)):
      print("Found lattitude values")
      print("lat :",pre_lat,"Index :",index1)
      
      #Check longitude
      if ((wind_long-20)<pre_long<(wind_long+20)):
        print("Found longitude values")
        pressure_array[index] = pressure_row['surface_pressure']
        humidity_array[index] = pressure_row['specific_humidity']
        temparature[index] = pressure_row['temperature']

        break
        
    else:
      print("In else")
  
print(pressure_array)
          
      
  
```
